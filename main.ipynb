Feature Description:
UTC - The time when experiment was performed.
Temperature - Temperature of Surroundings. Measured in Celsius
Humidity - The air humidity during the experiment.
TVOC - Total Volatile Organic Compounds. Measured in ppb (parts per billion)
eCo2 - CO2 equivalent concentration. Measured in ppm (parts per million)
Raw H2 - The amount of Raw Hydrogen present in the surroundings.
Raw Ethanol - The amount of Raw Ethanol present in the surroundings.
Pressure - Air pressure. Measured in hPa
PM1.0 - Paticulate matter of diameter less than 1.0 micrometer .
PM2.5 - Paticulate matter of diameter less than 2.5 micrometer.
NC0.5 - Concentration of particulate matter of diameter less than 0.5 micrometers.
NC1.0 - Concentration of particulate matter of diameter less than 1.0 micrometers.
NC2.5 - Concentration of particulate matter of diameter less than 2.5 micrometers.
CNT - Simple Count.
Fire Alarm - (Reality) If fire was present then value is 1 else it is 0.

```python
smoke.drop(['UTC','CNT'], axis = 1, inplace = True)
smoke
```

```python
X = smoke.copy()
X.drop('Fire Alarm',axis = 1,inplace = True)
y = smoke['Fire Alarm']
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3,random_state=10)
ss = StandardScaler() # Used coz iteration limit reached
X_train = ss.fit_transform(X_train)
X_test = ss.transform(X_test)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.dummy import DummyClassifier
from sklearn.naive_bayes import GaussianNB
import time

log_model=LogisticRegression()
log_model.fit(X_train,y_train)

begin1=time.time()
log_prediction = log_model.predict(X_test)
end1=time.time()
print((end1-begin1)*10**3,"ms")
```

```python
dum_model=DummyClassifier()
dum_model.fit(X_train,y_train)

begin2=time.time()
dum_prediction = dum_model.predict(X_test)
end2=time.time()
print((end2-begin2)*10**3,"ms")
```

```python
gua_model=GaussianNB()
gua_model.fit(X_train,y_train)

begin3=time.time()
gua_prediction = gua_model.predict(X_test)
end3=time.time()
print((end3-begin3)*10**3,"ms")
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, log_prediction)
# TN  FN
# FP  TP
```

```python
confusion_matrix(y_test, dum_prediction)
```

```python
confusion_matrix(y_test, gua_prediction)
```

```python
from sklearn.metrics import accuracy_score,precision_score
print(accuracy_score(log_prediction,y_test),precision_score(log_prediction,y_test)
)
```

```python
print(accuracy_score(dum_prediction,y_test),precision_score(dum_prediction,y_test))
```

```python
print(accuracy_score(gua_prediction,y_test),precision_score(gua_prediction,y_test))
```
